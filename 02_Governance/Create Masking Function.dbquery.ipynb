```sql
-- 1. Column Masking (Hide Customer Names)
CREATE OR REPLACE FUNCTION `ai-sales-insights-genie`.input_data.mask_customer(name STRING)
RETURN CASE 
    WHEN is_account_group_member('admin') THEN name 
    WHEN current_user() = 'vishnupachauri@learning.com' THEN name
    ELSE 'REDACTED' 
END;


ALTER TABLE `ai-sales-insights-genie`.input_data.fact_sales 
ALTER COLUMN CustomerName SET MASK `ai-sales-insights-genie`.input_data.mask_customer;


-- 2. Row Level Security (Advanced Hackathon Feature)
-- Only allow users to see data for their assigned State (Example logic)
CREATE OR REPLACE FUNCTION `ai-sales-insights-genie`.input_data.state_filter(state_name STRING)
RETURN CASE 
    WHEN is_account_group_member('admin') THEN TRUE 
    WHEN current_user() = 'vishnupachauri@learning.com' THEN TRUE
    WHEN state_name = 'Delhi' THEN TRUE -- Logic: Delhi users only see Delhi
    ELSE FALSE 
END;


ALTER TABLE `ai-sales-insights-genie`.input_data.fact_sales SET ROW FILTER `ai-sales-insights-genie`.input_data.state_filter ON (State);
```
